```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 6))


ax.plot(x, y, label='Sine Wave', color='b', linewidth=2)


ax.set_title('Customized Line Plot', fontsize=16, weight='bold')
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)
ax.grid(True, linestyle='--', color='gray', alpha=0.7)
ax.legend()


ax.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.3),
            arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 5, 8, 4]


fig, ax = plt.subplots(figsize=(8, 6))


bars = ax.bar(categories, values, color=['red', 'green', 'blue', 'orange', 'purple'])


ax.set_title('Customized Bar Chart', fontsize=16, weight='bold')
ax.set_xlabel('Categories', fontsize=12)
ax.set_ylabel('Values', fontsize=12)

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 2))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 12), sharex=True)

# Plotting the first subplot (Sine Wave)
ax1.plot(x, y1, label='Sine Wave', color='b', linewidth=2)
ax1.set_title('Sine Wave', fontsize=14)
ax1.set_ylabel('Amplitude', fontsize=12)
ax1.legend()

# Plotting the second subplot (Cosine Wave)
ax2.plot(x, y2, label='Cosine Wave', color='r', linewidth=2)
ax2.set_title('Cosine Wave', fontsize=14)
ax2.set_xlabel('X-axis', fontsize=12)
ax2.set_ylabel('Amplitude', fontsize=12)
ax2.legend()

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 100 * np.random.rand(100)  


fig, ax = plt.subplots(figsize=(8, 6))


scatter = ax.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')


ax.set_title('Customized Scatter Plot', fontsize=16, weight='bold')
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)
fig.colorbar(scatter, ax=ax, label='Color by Distance')


plt.show()
```
